```python
def wordcount(x):
  split = x.split(' ')
  return len(split)
```

```python
messages['word_count'] = messages['content_fix'].apply(wordcount)
```

```python
def countanh(x):    
    return x.count(' anh ')+x.count(' a ')+x.count('Anh')
def countem(x):
    return x.count(' em ')+x.count(' e ')+x.count('Em')+x.count('Emmm')+x.count('Emm')+x.count(' emm')+x.count(' emmm')
def countbac(x):
    return x.count(' bac ') + x.count(' bác ')
def countcau(x):
    return x.count('cậu')
def countto(x):
    return x.count(' tớ ')
def countnho(x):
    return x.count('nhớ')+x.count(' nho ')
def countyeuthich(x):
    return x.count('yeu')+x.count(' iu ')+x.count('yêu')+x.count('thich')+x.count('thích')
def countny(x):
    return x.count(' ny ')
def countsmile(x):
    return x.count(':)')+x.count('=)')
def countcry(x):
    return x.count(':(')+x.count('=(')
```

```python
messages['anh_count'] = messages['content_fix'].apply(countanh)
messages['em_count'] = messages['content_fix'].apply(countem)
messages['bac_count'] = messages['content_fix'].apply(countbac)
messages['cau_count'] = messages['content_fix'].apply(countcau)
messages['to_count'] = messages['content_fix'].apply(countto)
messages['ny_count'] = messages['content_fix'].apply(countny)
messages['yeuthich_cnt'] = messages['content_fix'].apply(countyeuthich)
messages['nho_cnt']= messages['content_fix'].apply(countnho)
messages['smile_cnt']=messages['content_fix'].apply(countsmile)
messages['cry_cnt']=messages['content_fix'].apply(countcry)
```
